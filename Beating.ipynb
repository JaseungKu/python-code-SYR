```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi

f0 = 10e6 - 200e3
f1 = 10e6
phi = 0/360 * 2 * pi
t0=5e-6

t = np.linspace(0, 5, 5000) * 1e-6
y0 = np.exp(-t/t0)*np.sin(2*pi*f0*t)
y1= np.sin(2*pi*f1*t + phi)

y2 = y0 + y1

plt.plot(t, y0, t, y1, t,y2)
```
