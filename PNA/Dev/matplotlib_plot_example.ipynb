```python
# Plot as fast as possible.
%matplotlib 
import numpy as np
import matplotlib.pyplot as plt
import time

xdata, ydata = [], []

fig, axes = plt.subplots(1,1)
lines, = axes.plot(xdata, ydata, 'b-') 
# or lines = plt.plot(xdata, ydata)[0]
# or lines, =axes.plot(xdata, ydata)

axes.set_xlabel('index') #plt.xlabel('index') works as well
plt.ylabel('arb')
plt.title('read sound')

i = 0
xdata = np.linspace(0,999,1000)
startTime = time.time()

while i<10:    # take data indefinitely
    
    ydata = np.random.normal(size=1000)
    
    lines.set_xdata(xdata)    
    lines.set_ydata(ydata)    
    
    axes.relim()
    axes.autoscale()
    fig.canvas.draw()  
    fig.canvas.flush_events()

    i += 1
    
#print 'fps = %.2f' % (100.0/(time.time()-startTime))
```

```python
%matplotlib
# With %matplotlib, a figure is newly created. The old one disappears.
import numpy as np
from numpy import newaxis
import matplotlib.pyplot as plt
import time

class plot2D(object):
    """
    Description:
    """
    def __init__(self, xlabel='xlabel()', ylabel='ylabel()', title='title', figsize=(10,5)):
        self.fig, self.axes = plt.subplots(nrows=1, figsize=figsize)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)
        plt.xticks    
        #self.axesimage = self.axes.matshow([[]])
        #plt.colorbar(self.axesimage)
        #self.textvar = plt.figtext(0.75, 0.92, 'variable=')
    
    def update(self, array_2D, extent, cmap='Blues_r'):
        self.axesimage = self.axes.matshow(array_2D, cmap=cmap, extent=extent, origin='lower', aspect='auto')
        self.axesimage.set_data(array_2D)
        self.axes.xaxis.set_ticks_position('bottom')  
        #plt.colorbar(self.axesimage)
        #plt.gcf().texts.remove(self.textvar) # This and next line make code slow. Use when appropriate.
        #self.textvar = plt.figtext(0.75, 0.92, 'variable='+str(var))
        self.axes.autoscale()
        plt.draw()
       
# setup plot
plotObj = plot2D(ylabel='S21 (dB)',xlabel='Frequency (Hz)', title='S21 vs Freq vs Flux', figsize=(8,4)) # create 2D plot.

# initialize data array
s21_2D = np.array([])

filename = 'test.txt'

i=0
while i<5:    
    
    s21_1D = np.random.rand(10) * 1e9
    
    if s21_2D.size == 0:
        s21_2D = s21_1D[:, newaxis] # make 2D array
    else:
        s21_2D = np.column_stack((s21_2D, s21_1D)) # update data.
    
    f = open(filename, 'bw') 
    np.savetxt(f, s21_2D, fmt='%.2f')
    f.close()
    
    extent = [-1, i, -2, 2]
    plotObj.update(s21_2D, extent, cmap='gray_r') # update plot
    time.sleep(0.2)
    i += 1
```
