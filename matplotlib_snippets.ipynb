```python
x = np.linspace(0,1,200)
y = np.sin(x**2*20)

fig, ax = plt.subplots(figsize=(10,7))
ax.plot(x, y)

ax.set_title('ax title', fontsize=24)

ax.set_xlabel('Frequency (GHz)', fontsize=24)
ax.set_ylabel('Amplitude (V)', fontsize=24)


# ax.set_xlim(0,0.1)
# ax.set_ylim(0,2)

# ax.set_xscale('log')
# ax.set_yscale('log')

ax.tick_params('both', labelsize=18)
ax.grid()
```
