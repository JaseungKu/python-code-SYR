# Two-transmons + cavity

```python
N, M =3,4
b1 = destroy(N)
b2 = destroy(N)
c = destroy(M)
Ib = qeye(N)
Ic = qeye(M)
Del1, Del2, del1, del2 = -0.9, -1, -0.3,-0.3
g1, g2 = 0.100, 0.100

# This gives Eq.(2.11) alike in Magesan paper.
H1 = Del1*tensor(b1.dag()*b1, Ib, Ic) + del1/2*tensor(b1.dag()*b1 * (b1.dag()*b1 - Ib), Ib, Ic)
H2 = Del2*tensor(Ib, b2.dag()*b2, Ic) + del2/2*tensor(Ib, b2.dag()*b2 * ( b2.dag()*b2 - Ib), Ic)
H3 = g1*(tensor(b1.dag(), Ib, c) + tensor(b1, Ib, c.dag())) + g2*(tensor(Ib, b2.dag(),  c) + tensor(Ib, b2, c.dag()))
H = H1 + H2 + H3

# This gives Eq.(2.6) in Magesan paper.
# H1 = Del1*tensor(Ic, b1.dag()*b1, Ib) + del1/2*tensor(Ic, b1.dag()*b1 * (b1.dag()*b1 - Ib), Ib)
# H2 = Del2*tensor(Ic, Ib, b2.dag()*b2) + del2/2*tensor(Ic,Ib, b2.dag()*b2 * ( b2.dag()*b2 - Ib))
# H3 = g1*(tensor(c, b1.dag(), Ib) + tensor(c.dag(), b1, Ib)) + g2*(tensor(c, Ib, b2.dag()) + tensor(c.dag(), Ib, b2))
# H = H1 + H2 + H3
```

# Two-level qubit + CR

```python
J = Symbol('J')
H_R = Matrix([[0,0,omega/2,0],[0,0,J,omega/2],[omega/2,J,delta,0],[0,omega/2,0,delta]])
H_R
```

```python
N1 = sqrt(2 + 2*omega**2/(2*delta**2+omega**2 + 2*delta*sqrt(delta**2+omega**2)))
N2 = sqrt(2 + 2*omega**2/(2*delta**2+omega**2 - 2*delta*sqrt(delta**2+omega**2)))
A1 = sqrt(2)/N1
A2 = sqrt(2)/N2
B1 = -sqrt(2)*omega/(delta + sqrt(delta**2+omega**2)) / N1
B2 = -sqrt(2)*omega/(delta - sqrt(delta**2+omega**2)) / N2

T = Matrix([[A1,0,A2,0]
           ,[0,A1,0,A2]
           ,[B1,0,B2,0]
           ,[0,B1,0,B2]])
T
```

```python
simplify(T)
```

```python
simplify(transpose(T)*H_R*T) 
# Digonal blocks match with Eq.3.12 and 3.13, but off-diagonal blocks are not zero, why?....
```
