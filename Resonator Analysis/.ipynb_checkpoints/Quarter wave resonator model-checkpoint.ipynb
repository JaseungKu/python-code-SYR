# Quarter wave resonator Fit

Ref : Mazin and Jao Ph.D thesis / Megrant et al. APL 100, 113510 (2012)/ Wisbey, J. APPLIED PHYSICS 108, 093918 (2010)

## S21 (dB) vs freq fit
Ref : Mazin Ph.D thesis
Note: Strictly speaking, this fit works only when $\phi_0=0$ in the s21 expression in Megrant and Wisbey paper.
\begin{eqnarray}
S_{21}&=&\frac{S_{21}^{min}+2iQ_L\delta x}{1+2iQ_L \delta x} \\
S_{21}^{min} &=& \frac{Q_c}{Q_i + Q_c} = \frac{1}{1+g} \\
g &=& \frac{Q_i}{Q_c},
\end{eqnarray}
where $Q_L, Q_i, Q_c$ : loaded, internal and coupling quality factor, and $\delta x = (f-f_0)/f_0$. 

The fitting functino should be of form,
$$S_{21}=A\frac{S_{21}^{min}+2iQ_L\delta x}{1+2iQ_L \delta x},$$
where $A$ is a scaling factor that accounts for uncalibrated transmission.
After replacing $Q_L =f_0/\Delta f$ and $\delta x=(f-f_0)/f_0$, we get 
$$S_{21} (dB)=20\log(|S_{21}|)= 20\log(A)+ 10\log\left(\frac{(S_{21}^{min})^2+4 (f-f_0)/\Delta f^2}{1++4 (f-f_0)/\Delta f^2}\right)$$
Therefore, we need four fitting parameters: $A$, $f_0$, $\Delta f$, and $S_{21}^{min}$.

Then using the fitted parameters $S_{21}^{min}$ and $Q_L=f_0/\Delta f$, we can get
\begin{eqnarray*}
g &=& \frac{1}{S_{21}^{min}} - 1\\
Q_i &=& (g+1)Q_L \\
Q_c &=& \left(\frac{g+1}{g}\right)Q_L \\
\end{eqnarray*}
## Geometric fit and four parameter fit
### Ref: Wisbey paper

$S_{21}$ is given by $$S_{21} = a\exp(-2\pi if\tau) \left(1 -\frac{(Q_L/Q_c)e^{i\phi_0}}{1+2iQ_L\delta x}\right)$$
The normalized transmission is given by 
$$S_{21} = 1 -\frac{(Q_L/Q_c)e^{i\phi_0}}{1+2iQ_L\delta x}$$
$S_{21}$ can be rewritten by
\begin{eqnarray}
S_{21}-1+\frac{Q_L}{2Q_c}e^{i\phi_0} &=& \frac{Q_L}{2Q_c}e^{i\phi_0}\left(\frac{-1+2iQ_L\delta x}{1+2iQ_L\delta x}\right) \\
\left|S_{21}-\left(1-\frac{Q_L}{2Q_c}e^{i\phi_0}\right) \right| &=& \frac{Q_L}{2Q_c}
\end{eqnarray}
The last equation means that when $S_{21}$ is plotted in complex plane, $S_{21}$ traces a circle with a diameter $D$ centerd at the point $C$. 
\begin{eqnarray}
D&=&\frac{Q_L}{Q_c}=\frac{g}{(1+g)} \\
C&=&1-\frac{Q_L}{2Q_c}e^{i\phi_0} 
\end{eqnarray}
Note that $0 > D > 1$ and the more overcoupled, $D\rightarrow1$ and the more undercoupled, $D\rightarrow0$.

Similarly, the inverse $S_{21}$ can be givey by,
$$S_{21}^{-1} = 1 +\frac{(Q_L/Q_c)e^{i\phi_0}}{(1-(Q_L/Q_c)e^{i\phi_0})+2iQ_L\delta x}$$

$S_{21}^{-1}$ traces a circle with a diameter $D$, centered at $C$.
\begin{eqnarray}
D&=&\left(\frac{Q_L}{Q_c}\right)\left(\frac{1}{1-(Q_L/Q_c)\cos\phi_0}\right)  \\
C&=&\left(1-\frac{Q_L}{2Q_c}e^{-i\phi0}\right)\left(\frac{1}{1-(Q_L/Q_c)\cos\phi_0}\right)
\end{eqnarray}

Note that unllike the $S_{21}$, the diameter $D$ of $S_{21}^{-1}$ depends on $\phi_0$.

As a special case of $\phi_0=0$,

\begin{eqnarray}
S_{21}^{-1} &=& 1 +\frac{(Q_L/Q_c)}{(1-(Q_L/Q_c))+2iQ_L\delta x} \\
S_{21}^{-1} -(1+\frac{Q_i}{2Q_c})&=&\frac{Q_i}{2Q_c}\left(\frac{Q_L/Q_i-2i\delta x Q_L}{Q_L/Q_i +2i\delta x Q_L}\right) \\
\left|S_{21}^{-1} -(1+\frac{Q_i}{2Q_c})\right|&=&\frac{Q_i}{2Q_c}
\end{eqnarray}

$S_{21}^{-1}$ traces a circle with a diameter $D=Q_i/Q_c$, centerd at the point $C=1+Q_i/(2Q_c)$.

#### Note
1. In S21 (dB) vs freq, the $Q_L$ is NOT $f_0/\Delta f_{3dB}$. The $Q_L$ should be calculated in |S21| vs freq. The $\Delta f$ is the difference of the frequencies where $|S_{21}|= 1- \left( \frac{1 -|S_{21}|^{min}}{\sqrt{2}}\right)$.
2. In general, $\Delta f$ obtained from two opposite points on the inverse s21 circle is NOT $\Delta f_{3dB}$. Also,  $$\frac{\Delta f}{f_0}=\frac{1}{Q_i}+\frac{1}{Q_c}\left(1-\sqrt{2}\sin(\phi_0 +\pi/4)\right)$$ If $\phi_0=0$,  $\Delta f/f_0 = 1/Q_i$.
3. It is not clear where $\phi_0$ comes from in the derivation of $S_{21}$. In fact, this $\phi_0$ is NOT the same as the $\phi$ in the Megrant pape, even thought $\phi_0$ is the rotation angle of the center of the $S_{21}^{-1}$ circle with respect to (1,0) point. One evidence is that according to the expression D above, D gets smaller as $\phi$ increases.Also, D must be less than or equal to $1/(1-\cos\phi)$, so there is a maximum in D (D_max), given $\phi$. However, I have seen data having much larger D than D_max. The term $e^{i\phi_0}$ might be there to account for the rotation angle of S21 circle.

### Ref: Megrant paper
This paper seems to present better model when it comes to getting Qi and Qc.
$$S_{21}^{-1} = 1 +\frac{(Q_i/Q_c^*)e^{i\phi}}{(1+2iQ_L\delta x}$$ 
,where $Q_c^*=(Z_0/|Z|)Q_c$.
The invers S21 makes a circle centered at C with diameter D:
$$ C=1+ \frac{D}{2}e^{i\phi},\,\,\, D = \frac{Qi}{Q_C^*}$$
The frequency difference $\Delta f$ of two points at $\pm$90 from the $f_0$ point is the linewith (or 3dB point) of ($S_{21}^{-1}$-1) and $Q_i$ is given by,
$$Q_i = \frac{f_0}{\Delta f}$$
By fitting the inverse S21 circle and finding D, $f_0$ and $\Delta f$, we can calculate $Q_i=f_0/\Delta f$ and $Q_c^*=(Z_0/|Z|)Q_c$. Therefore, $Q_i$ can be obtained accurately, but $Q_C$ can't. However, we know |Z| is pretty close to 50 $\Omega$, so we may assume $Q_c = Q_C^*$

#### Fit
Fit steps:
1. Normalize S21
Method1: Background substration on magnitude and phase of S21, so that the inverse S21 crosses (1,0) point and data converges to that point when $|\delta x|$ increases.
Method2: Mutiply $Ae^{i\theta}$ to S21 and find $A$ and $\theta$, so that the S21 and inverse S21 meet at (1,0) and the centers of S21 and inverse S21 and (1,0) are collinear.
2. Draw the inverse S21 and fit a circle to the data. 
3. Obtain the D and $\Delta f$ from the geometric fit. From them, calculate $Q_L$, $Q_i$ and $Q_c$.
    

## S21, S21 circle and inverse S21 circle based on Wisbey paper.

```python
# S21 from Wisbey paper

%matplotlib
import numpy as np
import matplotlib.pyplot as plt
 
# parameters
Qi = 1.7e6#100e3
Qc = 1e5 #50e3
f0 = 6.121e9
BW = 2e5
phi0 = 11.7/ 360 * 2 *np.pi

QL = Qi * Qc / (Qi + Qc)
print('QL={}, BW={}, S21_min={}'.format(QL, f0/QL, 20*np.log10(10)))

# create freq, S21 and S21^-1 array
f = np.linspace(f0-BW/2, f0+BW/2, 100000) #freq
A = 1 * np.exp(-1j* 2 * np.pi * f * 0e-9) 
s21 = A * (1 - (QL/Qc)*np.exp(1j*phi0)/(1+2j*QL*(f-f0)/f0))
s21_mag = np.absolute(s21)
s21_dB = 20 * np.log10(np.absolute(s21))
s21_phase = np.angle(s21, deg=1)
s21_inv = 1 / s21

# s21, s21^-2 circle
x, y = np.real(s21), np.imag(s21)
x_inv, y_inv = np.real(s21_inv), np.imag(s21_inv)

# divide into segments. boundary is 3dB points. 1: within 3dB, 2: out of 3dB
index1 = np.min(s21_dB) + 3 >=s21_dB
index2 = s21_dB > np.min(s21_dB) + 3
#index1 = 1- (1- np.min(s21_mag)) / np.sqrt(2) >= s21_mag
#index2 = s21_mag > 1- (1- np.min(s21_mag)) / np.sqrt(2) 
f1 = f[index1]
f2 = f[index2]

s21_1 = s21[index1]
s21_2 = s21[index2]

s21_mag1 = s21_mag[index1]
s21_mag2 = s21_mag[index2]

s21_phase1 = s21_phase[index1]
s21_phase2 = s21_phase[index2]

s21_dB1 = s21_dB[index1]
s21_dB2 = s21_dB[index2]

s21_inv1 = s21_inv[index1]
s21_inv2 = s21_inv[index2]

x1, y1 = np.real(s21_1), np.imag(s21_1)
x2, y2 = np.real(s21_2), np.imag(s21_2)
x_inv1, y_inv1 = np.real(s21_inv1), np.imag(s21_inv1)
x_inv2, y_inv2 = np.real(s21_inv2), np.imag(s21_inv2)

# center points
C1 = (1 - QL/(2*Qc)*np.exp(1j*phi0)) # center of S21 circle
C2 = (1 - QL/(2*Qc)*np.exp(-1j*phi0)) / (1-QL/Qc*np.cos(phi0)) # center of S21 inverse circle
C1_x, C1_y = np.real(C1), np.imag(C1)
C2_x, C2_y = np.real(C2), np.imag(C2)

# tracking point
i = 2790 # index of f

# plot
fig = plt.figure(figsize=(15,12))
fig.suptitle('Qi={:.3g}, Qc={:.3g}, QL={:.3g}'.format(Qi, Qc, QL), fontsize=14)
axes1 = fig.add_subplot(3,2,1)
axes2 = fig.add_subplot(3,2,3)
axes3 = fig.add_subplot(3,2,5)
axes4 = fig.add_subplot(2,2,2)
axes5 = fig.add_subplot(2,2,4)

axes1.plot(f1, s21_mag1, 'r.', f2, s21_mag2, 'k.')
axes1.grid()
axes1.set_ylabel('|S21|')
axes1.ticklabel_format(useOffset=False)

axes2.plot(f1, s21_dB1, 'r.', f2, s21_dB2, 'k.')
axes2.grid()
axes2.set_ylabel('S21 (dB)')
axes2.ticklabel_format(useOffset=False)

axes3.plot(f1, s21_phase1, 'r.', f2, s21_phase2, 'k.')
axes3.grid()
axes3.set_xlabel('Frequency (Hz)')
axes3.set_ylabel('Phase (degree)')
axes3.ticklabel_format(useOffset=False)

axes4.plot(x1, y1,'r.', x2, y2, 'k.', C1_x, C1_y, 'r-o',1, 0,'b-o')
axes4.grid()
axes4.set_xlabel('real($S_{21}$)')
axes4.set_ylabel('imag($S_{21}$)')
axes4.set_aspect('equal', adjustable='box')

axes5.plot(x1, y1,'r.', x2, y2, 'k.', x_inv1, y_inv1, 'r.', x_inv2, y_inv2, 'k.'
           , C1_x, C1_y, 'r-o', C2_x, C2_y, 'r-o', 1, 0,'b-o')
axes5.grid()
axes5.set_xlabel('real($S_{21}$, $S_{21}^{-1}$)')
axes5.set_ylabel('imag($S_{21}$, $S_{21}^{-1}$)')
axes5.set_aspect('equal', adjustable='box')
```

## S21, S21 circle and inverse S21 circle based on Megrant paper

```python
# S21 from Megrant paper
# Assume Qc*=Qc i.e. |Z|=50 Ohm

%matplotlib
import numpy as np
import matplotlib.pyplot as plt
 
# parameters
Qi = 1.7e6#100e3
Qc = 1e5 #50e3
f0 = 6.121e9
BW = 2e5
phi = 11.7/ 360 * 2 *np.pi

QL = Qi * Qc / (Qi + Qc)
print('QL={}, BW={}, S21_min={}'.format(QL, f0/QL, 20*np.log10(10)))

# create freq, S21 and S21^-1 array
f = np.linspace(f0-BW * 8, f0+BW *8, 100000) #freq

s21 =1/  (1 + (Qi/Qc)*np.exp(1j*phi)/(1+2j*QL*(f-f0)/f0) )
s21_mag = np.absolute(s21)
s21_dB = 20 * np.log10(np.absolute(s21))
s21_phase = np.angle(s21, deg=1)
s21_inv = 1 / s21

# s21, s21^-2 circle
x, y = np.real(s21), np.imag(s21)
x_inv, y_inv = np.real(s21_inv), np.imag(s21_inv)

# divide into segments. boundary is 3dB points. 1: within 3dB, 2: out of 3dB
index1 = np.min(s21_dB) + 3 >=s21_dB
index2 = s21_dB > np.min(s21_dB) + 3
#index1 = 1- (1- np.min(s21_mag)) / np.sqrt(2) >= s21_mag
#index2 = s21_mag > 1- (1- np.min(s21_mag)) / np.sqrt(2) 
f1 = f[index1]
f2 = f[index2]

s21_1 = s21[index1]
s21_2 = s21[index2]

s21_mag1 = s21_mag[index1]
s21_mag2 = s21_mag[index2]

s21_phase1 = s21_phase[index1]
s21_phase2 = s21_phase[index2]

s21_dB1 = s21_dB[index1]
s21_dB2 = s21_dB[index2]

s21_inv1 = s21_inv[index1]
s21_inv2 = s21_inv[index2]

x1, y1 = np.real(s21_1), np.imag(s21_1)
x2, y2 = np.real(s21_2), np.imag(s21_2)
x_inv1, y_inv1 = np.real(s21_inv1), np.imag(s21_inv1)
x_inv2, y_inv2 = np.real(s21_inv2), np.imag(s21_inv2)

# center points
C2 = 1 + Qi/Qc/2*np.exp(1j*phi) # center of S21 inverse circle

C2_x, C2_y = np.real(C2), np.imag(C2)

# tracking point
i = 2790 # index of f

# plot
fig = plt.figure(figsize=(15,12))
fig.suptitle('Qi={:.3g}, Qc={:.3g}, QL={:.3g}'.format(Qi, Qc, QL), fontsize=14)
axes1 = fig.add_subplot(3,2,1)
axes2 = fig.add_subplot(3,2,3)
axes3 = fig.add_subplot(3,2,5)
axes4 = fig.add_subplot(2,2,2)
axes5 = fig.add_subplot(2,2,4)

axes1.plot(f1, s21_mag1, 'r.', f2, s21_mag2, 'k.')
axes1.grid()
axes1.set_ylabel('|S21|')
axes1.ticklabel_format(useOffset=False)

axes2.plot(f1, s21_dB1, 'r.', f2, s21_dB2, 'k.')
axes2.grid()
axes2.set_ylabel('S21 (dB)')
axes2.ticklabel_format(useOffset=False)

axes3.plot(f1, s21_phase1, 'r.', f2, s21_phase2, 'k.')
axes3.grid()
axes3.set_xlabel('Frequency (Hz)')
axes3.set_ylabel('Phase (degree)')
axes3.ticklabel_format(useOffset=False)

axes4.plot(x1, y1,'r.', x2, y2, 'k.',1, 0,'b-o')
axes4.grid()
axes4.set_xlabel('real($S_{21}$)')
axes4.set_ylabel('imag($S_{21}$)')
axes4.set_aspect('equal', adjustable='box')

axes5.plot(x1, y1,'r.', x2, y2, 'k.', x_inv1, y_inv1, 'r.', x_inv2, y_inv2, 'k.'
           , C1_x, C1_y, 'r-o', C2_x, C2_y, 'r-o', 1, 0,'b-o')
axes5.grid()
axes5.set_xlabel('real($S_{21}$, $S_{21}^{-1}$)')
axes5.set_ylabel('imag($S_{21}$, $S_{21}^{-1}$)')
axes5.set_aspect('equal', adjustable='box')
```

## Transmission of quarter waver resonator in linear and log scale

```python
%matplotlib
import numpy as np
import matplotlib.pyplot as plt

class quarterWaveResonator:
    def __init__(self, Qi=1e4, Qc=1e4, f0=6e9, span=1e6, phi0=0, fnum=1e4):
        self.Qi, self.Qc = Qi, Qc    
        self.f0, self.phi0  = f0, phi0
        self.Ql = self.Qc * self.Qi / (self.Qi + self.Qc)
        self.span = span
        self.fnum = fnum
        self.freq = np.linspace(self.f0 - self.span/2, self.f0 + self.span/2, self.fnum)
        self.s21 =  1 - (self.Ql/self.Qc)*np.exp(1j*self.phi0)/(1+2j*self.Ql*(self.freq-self.f0)/self.f0)
        #self.s21 =  (self.Qc/(self.Qc+self.Qi)+2j * self.Ql*(self.freq-self.f0)/self.f0) /(1+2j*self.Ql*(self.freq-self.f0)/self.f0)
        self.s21_mag = np.absolute(self.s21)
        self.s21_dB = 20 * np.log10(np.absolute(self.s21))
        self.s21_phase = np.angle(self.s21, deg=1)
        self.s21_inv = 1 / self.s21
        self.s21_inv_mag = np.absolute(self.s21_inv)
        self.s21_inv_dB = 20 * np.log10(np.absolute(self.s21_inv))

#main
Qc_list = [1e2, 1e3, 1e4]
legend = []
fig = plt.figure(figsize=(17,6))
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.grid()
axes1.set_ylabel('|S21|')
axes1.ticklabel_format(useOffset=False)
axes2.grid()
axes2.set_ylabel('S21 (dB)')
axes2.ticklabel_format(useOffset=False)
                        
for q in Qc_list:
    res = quarterWaveResonator(Qi = 1e6, Qc = q, span=20e6)
    axes1.plot(res.freq, res.s21_mag)
    axes2.plot(res.freq, res.s21_dB)
    legend.append('Qc='+str(q))

axes1.legend(legend, loc='lower left')
axes2.legend(legend, loc='lower left')
fig.suptitle('Qi={:.3g}'.format(res.Qi), fontsize=15)

Qi_list = [1e4, 1e5, 1e6]
legend = []
fig = plt.figure(figsize=(17,6))
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.grid()
axes1.set_ylabel('|S21|')
axes1.ticklabel_format(useOffset=False)
axes2.grid()
axes2.set_ylabel('S21 (dB)')
axes2.ticklabel_format(useOffset=False)

for q in Qi_list:
    res = quarterWaveResonator(Qi = q, Qc = 1e5, span=2e6)
    axes1.plot(res.freq, res.s21_mag)
    axes2.plot(res.freq, res.s21_dB)
    legend.append('Qi='+str(q))
    
axes1.legend(legend, loc='lower left')
axes2.legend(legend, loc='lower left')
fig.suptitle('Qc={:.3g}'.format(res.Qc), fontsize=15)
```
